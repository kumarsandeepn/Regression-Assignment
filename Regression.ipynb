```python
## 1. What is Simple Linear Regression?


# Simple linear regression is a statistical method used to model the relationship between one independent variable (also known as a predictor or explanatory variable) and one dependent variable (also known as a response or outcome variable) using a straight line. It assumes a linear relationship between the two variables, meaning the change in the dependent variable is proportional to the change in the independent variable. 

Example:Predict a student's exam score based on hours studied.

```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
# Hours studied (X) and corresponding exam scores (Y)
X = np.array([[1], [2], [3], [4], [5]])   # independent variable (2D)
y = np.array([50, 55, 65, 70, 75])        # dependent variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict scores
y_pred = model.predict(X)

# Print coefficients
print("Intercept (Œ≤‚ÇÄ):", model.intercept_)
print("Slope (Œ≤‚ÇÅ):", model.coef_[0])

# Plot the data and regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. What are the key assumptions of Simple Linear Regression?

## The key assumptions of simple linear regression are linearity, independence, homoscedasticity, and normality. These assumptions ensure that the model accurately reflects the relationship between the independent and dependent variables and that the results are reliable. 
Here's a breakdown of each assumption:

1. Linearity:
A linear relationship exists between the independent variable (X) and the dependent variable (Y). In other words, the relationship can be approximated by a straight line. 

2. Independence:
The residuals (the differences between observed and predicted values) are independent of each other. This means that the error for one observation does not influence the error for any other observation. 

3. Homoscedasticity:
The variance of the residuals is constant across all levels of the independent variable. In simpler terms, the spread of the data points around the regression line should be roughly the same for all values of X. 

4. Normality:
For any given value of X, the dependent variable Y is normally distributed. This means that the distribution of the errors is also normal. 
These assumptions are crucial for the validity of the linear regression model. If these assumptions are violated, the model's predictions may be inaccurate, and the statistical tests performed on the model may not be reliable.

```python
## 5. How do we calculate the slope m in Simple Linear Regression?

##  In simple linear regression, the slope (often denoted as 'm' or Œ≤‚ÇÅ) is calculated using the formula: m = Œ£((x·µ¢ - xÃÑ)(y·µ¢ - »≥)) / Œ£((x·µ¢ - xÃÑ)¬≤), where (x·µ¢, y·µ¢) are the data points, xÃÑ is the mean of x values, and »≥ is the mean of y values. This formula essentially represents the covariance of x and y divided by the variance of x, reflecting how much y changes for a unit change in x. 

Detailed Explanation:

1. Understanding the components:
Œ£ represents the summation of all data points. 
x·µ¢ and y·µ¢ are individual data points (x, y) in your dataset. 
xÃÑ (x-bar) is the average of all x values: Œ£x·µ¢ / n, where n is the number of data points. 
»≥ (y-bar) is the average of all y values: Œ£y·µ¢ / n.


* (x·µ¢ - xÃÑ) and (y·µ¢ - »≥) represent the deviations of each x and y value from their respective means. 
Multiplying these deviations (covariance) indicates the direction and strength of the relationship between x and y. If they tend to be both above or below their means, the product will be positive, indicating a positive correlation; if one is above and the other below, the product will be negative, indicating a negative correlation. 
The denominator, Œ£((x·µ¢ - xÃÑ)¬≤), represents the variance of x. This value ensures that the slope is scaled appropriately. 
```

```python
## 10. What are the key assumptions of Multiple Linear Regression?

## Multiple linear regression relies on several key assumptions for accurate and reliable results. These include linearity, independence of errors, homoscedasticity, and normality of residuals. Additionally, multicollinearity (high correlation between independent variables) should be avoided. 

breakdown of these assumptions:

Linearity:
The relationship between the dependent variable and each independent variable should be linear. This means that a change in an independent variable corresponds to a constant change in the dependent variable. 
Independence of Errors:
The errors (residuals) should be independent of each other. This means that the error for one observation should not be related to the error for any other observation. This assumption is often violated in time series data, leading to autocorrelation. 
Homoscedasticity:
The variance of the errors should be constant across all levels of the independent variables. In simpler terms, the spread of the data points around the regression line should be roughly the same for all values of the independent variables. 
Normality of Residuals:
The residuals should be approximately normally distributed. This assumption is important for hypothesis testing and confidence interval calculations. 
No Multicollinearity:
The independent variables should not be highly correlated with each other. High multicollinearity can make it difficult to determine the individual impact of each independent variable on the dependent variable and can lead to unstable coefficient estimates. 

   
```

```python
## 11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

## heteroscedasticity is a standard linear regression model, it's assumed that the spread or variability of the error terms is consistent across all values of the independent variables. 

Heteroscedasticity occurs when this assumption is violated, meaning the error terms have different variances for different values of the independent variables. 

This can manifest as a "fan" or "cone" shape on a residual plot, where the spread of residuals increases or decreases as the predicted values change.


How Heteroscedasticity Affects Regression Results:

1. Inefficient Parameter Estimates:
Heteroscedasticity causes the standard errors of the regression coefficients to be underestimated. This leads to wider confidence intervals and makes it more likely to incorrectly reject the null hypothesis (Type I error).

2. Biased Standard Errors:
The estimated standard errors are not accurate representations of the true variability of the coefficients. This can lead to inaccurate hypothesis tests and confidence intervals.

3. Unreliable Hypothesis Tests:
Since the standard errors are biased, the t-statistics and p-values derived from them will be unreliable. This can lead to incorrect conclusions about the significance of the independent variables.

4. Poor Predictions:
Although heteroscedasticity does not cause bias in the coefficient estimates themselves, the inefficiency of the estimates can lead to less precise predictions. 

* Heteroscedasticity, in the context of Multiple Linear Regression, refers to the unequal variance of the error terms (residuals) across different levels of the independent variables. This violates a key assumption of Ordinary Least Squares (OLS) regression, which assumes homoscedasticity (constant variance of residuals). Heteroscedasticity can lead to unreliable and inefficient parameter estimates, making it difficult to interpret the results of the regression model. 
```

```python
## 12. How can you improve a Multiple Linear Regression model with high multicollinearity?

## High multicollinearity, where independent variables in a multiple linear regression model are highly correlated, can be addressed by several techniques. These include removing or combining correlated variables, using regularization methods like Ridge or Lasso regression, increasing the sample size, or employing dimensionality reduction techniques like Principal Component Analysis (PCA). 

Here's a more detailed look at each approach:

1. Removing or Combining Correlated Variables: 
Identify highly correlated variables:
Calculate the Variance Inflation Factor (VIF) for each predictor variable. A VIF value above 5 or 10 (depending on the context) indicates high multicollinearity. 
Remove redundant variables:
Drop one or more of the highly correlated variables, especially those with the highest VIFs, or those that are not essential to the model. 
Combine variables:
If variables are measuring similar aspects, create a new variable that combines them (e.g., combining income from salary and income from business into total income). 

2. Regularization Techniques:
Ridge Regression:
Adds a penalty term to the regression equation that is proportional to the square of the magnitude of the coefficients. This shrinks the coefficients, reducing the impact of multicollinearity.
Lasso Regression:
Similar to Ridge, but it can set some coefficients to zero, effectively removing those variables from the model. 

3. Increasing Sample Size:
More data helps: A larger dataset can help reduce the variance of the coefficient estimates, making the model more robust to multicollinearity. 

4. Dimensionality Reduction:
Principal Component Analysis (PCA): Transforms the original correlated variables into a smaller set of uncorrelated variables (principal components). This allows you to use the principal components as predictors instead of the original variables, effectively eliminating multicollinearity. 

5. Other Considerations:
Centering the data:
Subtracting the mean of each variable can sometimes improve the interpretation of coefficients, but it doesn't always eliminate multicollinearity. 

Stepwise Selection:
Iteratively adding or removing variables based on their statistical significance can help identify a subset of variables that minimize multicollinearity. 

    Understanding the cause:
If the multicollinearity is due to variables measuring the same underlying construct, consider whether one of those variables can be removed based on the context of the research or analysis. 
```

```python
## 13. What are some common techniques for transforming categorical variables for use in regression models?

## To incorporate categorical variables into regression models, common techniques include one-hot encoding, label encoding, and dummy variable creation. One-hot encoding creates binary columns for each category, while label encoding assigns a unique integer to each category. Dummy variable creation, a form of one-hot encoding, represents categories with binary variables, with one category typically omitted as a reference. 

1. One-Hot Encoding:
This method creates a new binary variable (0 or 1) for each unique category in the original categorical variable.
For example, if you have a "color" variable with categories "red", "blue", and "green", you would create three new variables: "color_red", "color_blue", and "color_green".
A data point with "red" would have a 1 in the "color_red" column and 0s in the other two.
This is widely used because it avoids imposing any arbitrary order on the categories and is suitable for many regression algorithms. 

2. Label Encoding:
This technique assigns a unique integer to each category.
For instance, "red" could be 1, "blue" 2, and "green" 3.
While simple, this method can be problematic if the categories have no inherent order, as the model might interpret the numerical assignments as relationships.
It is often suitable for ordinal data (where categories have a meaningful order). 

3. Dummy Variable Creation:
Similar to one-hot encoding, dummy variables are created to represent each category.
However, one category is typically omitted and used as a reference (e.g., the "baseline" or "control" group).
For example, if you have a "city" variable with categories "New York", "Los Angeles", and "Chicago", you might create dummy variables for "New York" and "Los Angeles", using "Chicago" as the reference.
The coefficients for the dummy variables then represent the difference in the outcome variable compared to the reference category. 

4. Other Techniques:
Target Encoding: Uses the average value of the target variable for each category as the encoding.
Frequency Encoding: Replaces each category with its frequency in the dataset.
Binning: Groups infrequent categories into a single "other" category before encoding. 

The best choice of technique depends on the specific dataset and the regression model being used. One-hot encoding and dummy variables are generally preferred for their interpretability and compatibility with many models. 
```

```python
## 16. What is the significance of the slope in regression analysis, and how does it affect predictions?

## Significance of the Slope:

Relationship Strength:
The slope's magnitude (absolute value) reflects the strength of the relationship between the variables. A larger slope (either positive or negative) indicates a stronger association, meaning changes in the independent variable have a greater impact on the dependent variable. 
Direction of Relationship:
The sign of the slope (+ or -) indicates the direction of the relationship. A positive slope means the dependent variable increases as the independent variable increases, while a negative slope indicates the dependent variable decreases as the independent variable increases. 

Predictive Power:
The slope is used to predict the value of the dependent variable for a given value of the independent variable. A larger slope allows for more accurate predictions, especially when extrapolating beyond the observed data. 
Hypothesis Testing:
The slope is a key component in hypothesis tests about the relationship between variables. Statistical tests (like t-tests) are used to determine if the observed slope is significantly different from zero, indicating a meaningful relationship. 
How the Slope Affects Predictions:
Accurate Predictions:
A well-defined slope, especially when based on a strong correlation (high r-squared value), allows for more accurate predictions of the dependent variable. 
Extrapolation:
The slope can be used to extrapolate (estimate values outside the range of observed data), but this should be done with caution, especially if the relationship is not linear. 
Outliers:
Outliers (unusually high or low values) can significantly affect the slope and therefore the accuracy of predictions. 
Model Interpretation:
The slope helps in interpreting the model's behavior. For example, in a study of weight and height, a slope of 0.3 means that for every additional inch of height, the weight is predicted to increase by 0.3 pounds. 
In simpler terms:
Imagine predicting how many followers you'll gain based on the number of posts you make. The slope is like a multiplier. A higher slope means each post has a greater impact on your follower count. If the slope is close to zero, your posts have little effect on followers. If the slope is negative, your posts are actually causing you to lose followers. 
```

```python
## 18. What are the limitations of using R¬≤ as a sole measure of model performance?

## Limitations of Using R¬≤ (R-squared) as the Sole Measure of Model Performance
While R¬≤ is a useful metric, relying on it alone can be misleading. 

1. R¬≤ Always Increases with More Predictors
Problem: R¬≤ can never decrease when you add more independent variables, even if they are irrelevant.

Why it matters: It may give a false impression that the model has improved.

Better: Use Adjusted R¬≤, which penalizes unnecessary predictors.

2. Does Not Indicate Causality
Problem: A high R¬≤ does not imply that X causes Y.

Example: Ice cream sales and drowning both rise in summer ‚Äî high correlation, but no causation.

3. Insensitive to Overfitting
Problem: A model might overfit the training data and still have a high R¬≤.

Why it matters: The model may perform poorly on unseen data.

4. Only Measures Linear Fit
Problem: R¬≤ assumes the relationship is linear.

If the true relationship is non-linear, R¬≤ may be low even for a good model.

5. Not Useful for Comparing Different Response Variables
You can‚Äôt compare R¬≤ across models with different dependent variables or units.

6. Ignores Residual Patterns
R¬≤ does not reveal whether residuals (errors) are randomly distributed.


```

```python
## 20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

## Heteroscedasticity, the unequal variance of errors in a regression model, is identifiable in residual plots as a fan or cone shape where the spread of residuals (errors) changes systematically with the predicted values. Addressing it is crucial because heteroscedasticity can lead to unreliable statistical inferences, making it more likely to incorrectly identify relationships as significant. 
Identifying Heteroscedasticity in Residual Plots:

Visual Inspection:
The most common method is to plot the residuals against the fitted (predicted) values. A random scattering of points with no discernible pattern suggests homoscedasticity (equal variance). A fan or cone shape, where the spread of residuals increases or decreases with the fitted values, indicates heteroscedasticity. 

Pattern Recognition:
Look for a systematic change in the spread of residuals. If the residuals are tightly clustered at some fitted values and widely scattered at others, it points to heteroscedasticity. 
Why Addressing Heteroscedasticity is Important:

Biased Coefficient Estimates:
Heteroscedasticity can lead to inaccurate estimates of the regression coefficients, making it difficult to determine the true effect of independent variables. 

Invalid Statistical Inferences:
The standard errors of the coefficients are underestimated, leading to incorrect conclusions about the statistical significance of variables. It becomes easier to falsely conclude that a variable is significant when it's not, or vice-versa. 

Poor Prediction Accuracy:
Heteroscedasticity can result in less accurate predictions, particularly for values in regions where the variance is high. 

Impacts on Economic Models:
In economic analysis, heteroscedasticity can lead to incorrect policy recommendations and suboptimal decisions. 
In essence, heteroscedasticity undermines the reliability of regression analysis and can lead to misleading conclusions. Therefore, it's crucial to identify and address it to ensure the validity and accuracy of the model and its interpretations. 
```

```python
## 21. What does it mean if a Multiple Linear Regression model has a high R¬≤ but low adjusted R¬≤?

##  A high R-squared with a low Adjusted R-squared in a multiple linear regression model suggests that the model includes many predictors, some of which may not be significantly contributing to explaining the variance in the dependent variable. Essentially, the model is likely overfitting the data, capturing noise along with the true signal. 

R-squared:
Measures the proportion of variance in the dependent variable explained by all the independent variables in the model. A high R-squared (closer to 1) indicates a good fit, meaning the model explains a large portion of the variance. 
Adjusted R-squared:
A modified version of R-squared that penalizes the addition of irrelevant predictors. It only increases when a new predictor improves the model's predictive power beyond what would be expected by chance. 
High R-squared, Low Adjusted R-squared:
This combination indicates that while the model as a whole explains a lot of the variance, many of the individual predictors are not adding much to the explanation. This is because the adjusted R-squared penalizes the inclusion of these less helpful predictors, leading to a lower value. 
Overfitting:
When a model includes too many predictors, it can start to fit the noise or random fluctuations in the specific training data rather than the underlying relationships. This results in a model that performs well on the training data but poorly on new, unseen data. 
Consequences:
A model with a high R-squared but low adjusted R-squared is likely not a good model for prediction, as its performance is not generalizable to new data.
```

```python
## 22. Why is it important to scale variables in Multiple Linear Regression?

##  Scaling variables in multiple linear regression is crucial for several reasons, primarily impacting the model's training process, performance, and interpretability. It ensures that features contribute equally to the model's calculations and can prevent certain features from dominating due to their larger scales. 

explanation:
1. Gradient Descent Convergence:
Many machine learning algorithms, including those used in multiple linear regression, rely on gradient descent for optimization. 
When features have vastly different scales, the gradients (which indicate the direction of the steepest ascent/descent) can become very large for some features and very small for others. 
This can cause gradient descent to oscillate or take a long time to converge to the optimal solution, or even prevent it from converging at all. 
Scaling the features to a similar range (e.g., between 0 and 1 or using standardization) ensures that the gradients are more balanced and that the algorithm can converge more efficiently. 

2. Impact on Model Coefficients:
In multiple linear regression, the model learns coefficients (weights) for each feature to predict the target variable. 
If some features have much larger scales than others, their corresponding coefficients can become very small or very large, making it difficult to interpret the relative importance of each feature. 
Scaling the data ensures that the coefficients reflect the true relative importance of each feature, making the model more interpretable. 

3. Numerical Stability:
Some calculations involved in multiple linear regression can be sensitive to the scale of the input data. 
Large differences in scale can lead to numerical instability, potentially causing errors or inaccurate results. 
Scaling the data helps to mitigate these issues and improve the overall stability of the model. 

4. Regularization:
Regularization techniques (like L1 or L2 regularization) are used to prevent overfitting by adding a penalty term to the model's loss function. 

These penalties are often applied to the coefficients of the model. 
If features have different scales, the regularization term will disproportionately affect features with larger scales. 
Scaling the data ensures that the regularization is applied fairly to all features, preventing one feature from dominating the regularization process.
```

```python
## 24. How does polynomial regression differ from linear regression?

## Linear regression uses a straight line to model the relationship between variables, while polynomial regression can use curved lines (polynomials) to model more complex, non-linear relationships. Essentially, polynomial regression extends linear regression by including higher-degree terms of the independent variable, allowing it to fit curves instead of just lines. 

Linear Regression: 
Assumes a linear relationship between the independent and dependent variables. 
Uses a straight line (y = mx + b) to model the relationship. 
Simple to implement and interpret. 
Suitable when data points appear to roughly follow a straight line on a scatter plot. 
Can be inaccurate if the relationship is non-linear. 

Polynomial Regression:
Can model non-linear relationships, allowing for curves in the data. 
Uses polynomial equations (e.g., y = ax¬≤ + bx + c) to fit the data. 
Provides more flexibility to capture complex patterns. 
Higher-degree polynomials can overfit the data (fit noise rather than the underlying trend). 
Requires careful selection of the polynomial degree to avoid overfitting. 
Can be more computationally expensive than linear regression. 
```

```python
## 26. What is the general equation for polynomial regression?

##  The general equation for polynomial regression of degree n is: y = b‚ÇÄ + b‚ÇÅx + b‚ÇÇx¬≤ + ... + b‚Çôx‚Åø + Œµ, where 'y' is the dependent variable, 'x' is the independent variable, 'b‚ÇÄ' is the y-intercept, 'b·µ¢' are the coefficients for each term, and 'Œµ' represents the error term. 

y: The predicted value of the dependent variable. 
x: The independent variable. 
b‚ÇÄ: The y-intercept, the point where the polynomial crosses the y-axis. 
b‚ÇÅ, b‚ÇÇ, ..., b‚Çô: The coefficients for each term in the polynomial. These values determine the shape and direction of the curve. 
x¬≤, x¬≥, ..., x‚Åø: The independent variable raised to different powers (degree of the polynomial). 
Œµ: Represents the error or residual, the difference between the predicted value and the actual value. 
```

```python
## 27. Can polynomial regression be applied to multiple variables?

## Yes, polynomial regression can be applied to multiple variables, also known as multivariate polynomial regression. This allows you to model relationships between multiple independent variables and a dependent variable, including non-linear relationships and interactions between the independent variables. 

1. Understanding Multivariate Polynomial Regression:
Multiple Variables:
Instead of just one independent variable (like in simple polynomial regression), you have several. For example, you might be predicting house prices based on square footage, number of bedrooms, and location.
Non-linear Relationships:
Polynomial regression allows you to model non-linear relationships between the independent variables and the dependent variable. This is done by including polynomial terms (like x¬≤, x¬≥, etc.) and interaction terms (like x*y) in the regression model.
Interaction Effects:
Polynomial regression can capture how the relationship between one independent variable and the dependent variable changes depending on the value of another independent variable. For example, the effect of adding another bedroom might depend on the size of the house. 

2. How it works:
You create new variables by taking powers and combinations of your original independent variables. For example, if your independent variables are x1 and x2, you might create x1¬≤, x2¬≤, and x1*x2. 
These new variables are then treated as additional independent variables in a multiple regression model. 
The regression coefficients for these new variables capture the non-linear and interaction effects. 

3. Why use it?
Improved Accuracy:
When relationships are not linear, polynomial regression can provide a better fit to the data than simple linear regression.
More Realistic Models:
In many real-world scenarios, relationships are not perfectly linear, and polynomial regression allows you to model more complex patterns.
```

```python
## 28. What are the limitations of polynomial regression?

##  Limitations of Polynomial Regression
While Polynomial Regression can model non-linear relationships, it comes with several important limitations that can affect its performance and reliability.

1. Overfitting
Problem: As the degree of the polynomial increases, the model may start fitting the noise in the data.

Result: Excellent performance on training data but poor generalization on unseen/test data.

Especially common when using high-degree polynomials.

2. Poor Extrapolation
Problem: Polynomial regression is unreliable for predicting values outside the range of the training data.

Why: The curve can behave erratically at the ends (especially with higher degrees).

Sensitive to Outliers
Problem: A single outlier can drastically change the fitted curve.

Why: Higher-degree polynomials try to pass through or near every point, including outliers.

Complexity Increases Quickly
Problem: As you increase the polynomial degree:

The number of terms grows

The model becomes harder to interpret

Computational cost increases

Multicollinearity Between Polynomial Terms
Problem: Polynomial features (like 
ùëã, X2, X3  are often highly correlated.

Effect: Leads to unstable coefficient estimates and inflated standard errors.

 Fix: Use regularization (Ridge or Lasso) to handle it.

6. May Not Capture True Relationship
Problem: If the underlying relationship isn't well approximated by a polynomial (e.g., exponential, logistic), polynomial regression may give misleading results.

```

```python
## 30. Why is visualization important in polynomial regression?

## Visualization plays a critical role in understanding, evaluating, and improving polynomial regression models. Here's why:

 1. Understand Non-Linear Relationships
Polynomial regression is used to model curved (non-linear) patterns.

Visualizing the curve helps confirm if the model is capturing the true trend in the data.

 Without visualization, it's hard to interpret complex polynomial curves.

 2. Detect Overfitting or Underfitting
Underfitting: The curve is too simple and misses the data pattern.

Overfitting: The curve is too complex and fits noise.

A visual plot makes it easy to spot both problems:



Too simple ‚Üí underfit
Too wiggly ‚Üí overfit
Just right ‚Üí good fit

3. Compare Model Performance Across Degrees
Plotting models with different degrees (1, 2, 3, etc.) on the same graph shows which one best fits the data.

Helps in selecting the best degree visually along with numerical metrics.

4. Explain Results to Others
A graph showing the data points and regression curve is easier to communicate to stakeholders who may not understand the math.

5. Check Extrapolation Issues
Visualization helps you see how the curve behaves outside the training range (left and right ends), where polynomial models can become unstable.

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.2, 1.9, 3.2, 3.9, 5.1])

# Fit polynomial of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = LinearRegression().fit(X_poly, y)

# Predict and plot
X_plot = np.linspace(0, 6, 100).reshape(-1, 1)
X_plot_poly = poly.transform(X_plot)
y_plot = model.predict(X_plot_poly)

plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X_plot, y_plot, color='red', label='Polynomial Fit (deg=2)')
plt.legend()
plt.title("Polynomial Regression Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Independent variable (feature)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)

# Dependent variable (target)
y = np.array([1, 4, 9, 16, 25, 36])  # y = x^2 + some noise

# Independent variable (feature)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)

# Dependent variable (target)
y = np.array([1, 4, 9, 16, 25, 36])  # y = x^2 + some noise


poly = PolynomialFeatures(degree=2)  # degree=2 for quadratic
X_poly = poly.fit_transform(X)


model = LinearRegression()
model.fit(X_poly, y)


X_test = np.linspace(1, 6, 100).reshape(-1, 1)
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)


plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X_test, y_pred, color='red', label='Polynomial Fit')
plt.title("Polynomial Regression (Degree = 2)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
